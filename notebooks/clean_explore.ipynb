```python
most_common = ['eggs', 'milk', 'chocolate', 'chocolate bars', 'large egg whites', 'dry yeast', 'corn starch', 'heavy cream', 'white sugar', 'vanilla beans', 'vanilla', 'vanilla extract', 'bread yeast', 'active dry yeast', 'instant yeast', 'plain flour', '2% reduced fat milk', 'bread', 'egg yolks', 'large egg yolks', 'whole milk', 'shortening', 'bread flour', 'cake flour', 'oil', 'large eggs', 'brown sugar', 'baking powder', 'sugar', 'coffee', 'ice water', 'all-purpose flour', 'cold water', 'hot water', 'warm water', 'granulated sugar', 'whipping cream', 'self rising flour', 'whole wheat flour', 'table salt', 'kosher salt', 'confectioners sugar', 'boiling water', 'extra-virgin olive oil', 'melted butter', 'garlic', 'lard', 'lemon', 'lemon juice', 'vegetable shortening', 'lime', 'sweetened condensed milk', 'olive oil', 'flour', 'all purpose unbleached flour', 'cocoa powder', 'condensed milk', 'sea salt', 'wheat', 'water', 'salt', 'yeast', 'butter', 'vegetable oil', 'unsalted butter']
```

```python
removed = 0
for recipe in tr:
    if len(recipe['ingredients']) <= 6:
        temp = recipe['ingredients']
        for ing in most_common:
            if ing in temp:
                temp.remove(ing)
        if len(temp) == 0:
            removed += 1
            tr.remove(recipe)
    if recipe['ingredients'] == [[]]:
        tr.remove(recipe)
        
print(f'removed: {removed}')
```
